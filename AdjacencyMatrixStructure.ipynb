# Adjacency Matrix Structure - Estructura de Matriz de Adyacencia

Realice lo siguiente:
  * Programe la representación de grafos mediante un la **Estructura de Matriz de Adyacencia**.
  * Mediante networkx muestre el resultado gráfico del gráfo.

```python
import networkx as nx
import numpy as np
```

```python
vertx = ['A', 'B', 'C', 'D']
matrix_adj = [[1, 0, 0, 2], 
              [5, 4, 0, 0],  
              [4, 2, 3, 0],  
              [1, 6, 8, 0]]
```

```python
G = nx.Graph()
```

```python
for i in range(len(vertx)):
    for j in range(len(vertx)):
        if matrix_adj[i][j] != 0:
            G.add_edge(vertx[i], vertx[j], weight=matrix_adj[i][j])
```

```python
nx.draw(G, with_labels=True,node_size=1000,font_color ='Green',font_weight = 'bold', font_size = 15, node_color='Black')
```
